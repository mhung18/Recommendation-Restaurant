```python
import pandas as pd
import itertools
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
data = pd.read_json('./restaurants.json')
data.head()
```

```python
resCategories = data['food_categories']
```

```python
# "mở phẳng" danh sách con trong mỗi hàng
all_items = list(itertools.chain.from_iterable(data["food_categories"]))

# lấy danh sách các giá trị duy nhất
categories = list(set(all_items))

print(categories)
```

```python
print(len(categories))
```

```python
def setCategoriesMatrix(catelist):
    resCatMatrix = []
    for x in categories:
        if (x in catelist):
            resCatMatrix.append(1)
        else:
            resCatMatrix.append(0)
    return resCatMatrix
```

```python
list_categories_matrix = [setCategoriesMatrix(x) for x in resCategories]
```

```python
for x in list_categories_matrix:
    print(x)
print(len(list_categories_matrix))
```

```python
cosine_sim = cosine_similarity(list_categories_matrix, list_categories_matrix)
# print(np.round(cosine_sim, 2).shape)
print(cosine_sim.shape)
print(cosine_sim.columns)
```

```python
feature = ['name','food_categories']
X = data[feature]

X.head()
```

```python
def recommend_similar_places(place_name, cosine_sim, df, top_n=5):
    # Kiểm tra quán có tồn tại trong DataFrame không
    if place_name not in df['name'].values:
        return f"Không tìm thấy quán '{place_name}' trong dữ liệu."

    # Tìm vị trí (index) của quán đó
    idx = df[df['name'] == place_name].index[0]

    # Lấy hàng tương ứng trong ma trận cosine
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sắp xếp theo độ tương đồng giảm dần (bỏ chính quán đó)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]

    # Lấy danh sách index quán tương tự
    similar_indices = [i[0] for i in sim_scores]

    # Trả về danh sách tên quán tương tự
    return df.iloc[similar_indices]['name'].tolist()
```

```python
print(recommend_similar_places('Bún Mắm Cô Ba - Lê Trọng Tấn',cosine_sim, X))
```
